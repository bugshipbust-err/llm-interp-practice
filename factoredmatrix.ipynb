```python
from transformer_lens import HookedTransformer, FactoredMatrix
```

```python
model = HookedTransformer.from_pretrained("tiny-stories-1M")
```

```python
W_in, W_out = model.blocks[7].mlp.W_in, model.blocks[7].mlp.W_out
```

```python
fact_mat = FactoredMatrix(W_in, W_out)
```

```python
fact_mat.BA.size(), fact_mat.AB.size()    # calc their product
```

```python
fact_mat.U.size(), fact_mat.S.size(), fact_mat.Vh.size()
```

```python
fact_mat.get_corner(k=3)   # returns a corner of fact_mat.AB
```

```python
len(fact_mat.svd())   # calculates svd efficiently
```

```python
fact_mat.norm()   # frobenius norm is sqrt(sum of squared singular values)
```
